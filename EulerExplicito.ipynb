```python
import numpy as np
from math import e
import matplotlib.pyplot as plt

t = 0
tf = 0.3
h = 0.1
n = round((tf - t)/h)
u = 2
dut = - u + t + 2

def aproximada(u,t,dut):
  valores = []
  for i in range(n):
    dut = dut - u + t + 2
    u = u + h*dut
    t = t + h
    valores.append(u)
  return valores

print(aproximada(u,t,dut))

def exata(t):
  valores = []
  for i in range(n):
    t = t + h
    u = t + 1 +(1/e**t)
    valores.append(u)
  return valores

print(exata(t))

# Lista da solução aproximada
valores_aproximados = aproximada(u, t, dut)

# Lista da solução exata
valores_exatos = exata(t)

# Plotagem dos gráficos
iteracoes = np.linspace(t, tf, n, endpoint=False)
plt.plot(iteracoes, valores_aproximados, 'r-', label='Solução Aproximada')
plt.plot(iteracoes, valores_exatos, 'b-', label='Solução Exata')
plt.legend()
plt.xlabel('Iteraçoes')
plt.ylabel('Valor de u')
plt.title('Comparação entre Solução Aproximada e Exata')
plt.show()
```
